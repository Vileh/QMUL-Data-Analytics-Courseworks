```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Data
link: https://www.kaggle.com/datasets/danmcinerney/mma-differentials-and-elo

```python
data = pd.read_csv('UFC_Data/masterdataframe.csv')

# Remove every fight before UFC 28 (17/11/2000) and all collumns after the 60th
# This is because UFC 28 was the first to use the unified rules of MMA
# The collumns after the 60th are differentials and averages of the first 60
data = data.iloc[464:, 0:60]

# Set date to datetime format
data.date = pd.to_datetime(data.date)

print(data.columns)

data.tail()
```

```python
data.describe()
```

```python
# Compare two fighters
fighter1 = 'Conor McGregor'
fighter2 = 'Khabib Nurmagomedov'
column = 'takedowns_attempts'

mcgregor_data = data.loc[data['fighter'] == fighter1][column].reset_index(drop=True)
khabib_data = data.loc[data['fighter'] == fighter2][column].reset_index(drop=True)

plt.plot(mcgregor_data, label=fighter1)
plt.plot(khabib_data, label=fighter2)
plt.legend()
```

```python
unique_fighters = data['fighter'].unique()
numerical_columns = data.select_dtypes(include=[np.number]).columns.tolist()
print(numerical_columns + ['fighter'])
# Create a dataframe with average stats of fighters
data[numerical_columns + ['fighter']].groupby('fighter').mean()
```

```python
data._get_numeric_data()
```

```python
# Plot the correlations
plt.figure(figsize=(15, 10))
data._get_numeric_data().corr()['result'].sort_values().drop('result').plot(kind='bar')
plt.show()
```

Notes: Accuracy is more important than attempts. Head strikes are more important than body strikes. It is very important to be good on the ground - both offensively and defensively. Clinchwork isn't that important.

```python
# Plot the correlations
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
data_1 = data.loc[data['total_comp_time'] == 900]._get_numeric_data().corr()['result'].sort_values().drop(['result', 'total_comp_time', 'round'])
data_2 = data.loc[data['total_comp_time'] == 1500]._get_numeric_data().corr()['result'].sort_values().drop(['result', 'total_comp_time', 'round'])
data_3 = data.loc[data['round'] == 1]._get_numeric_data().corr()['result'].sort_values().drop(['result', 'total_comp_time', 'round'])

data_2 = data_2.reindex(data_1.index, axis=1)
data_3 = data_3.reindex(data_1.index, axis=1)


# Calculate the width of each bar
width = 0.3

# Create an array with the positions of each group of bars
bar_positions = np.arange(len(data_1.index))

# Plot the bars
ax.bar(bar_positions - width, data_1.values, width, label='3 rounds decision', alpha=0.8)
ax.bar(bar_positions, data_2.values, width, label='5 rounds decision', alpha=0.8)
ax.bar(bar_positions + width, data_3.values, width, label='First round finishes', alpha=0.8)

# Set the x-ticks to be the names of the features
ax.set_xticks(bar_positions)
ax.set_xticklabels(data_1.index, ha='center')

ax.legend()
ax.xaxis.set_tick_params(rotation=90)
```
