### 2.1 Background

The broad goal of the coursework is to use the techniques we have learned in the course to study the solutions to a particular non linear partial differential equation called [the Sine-Gordon equation](https://en.wikipedia.org/wiki/Sine-Gordon_equation). 

Once complete, the notebook should provide a self contained report on your numerical investigations and provide relevant commentary in markdown cells around it that would enable another student to understand, run and make basic modifications to the notebook.

The Sine Gordon equation is as follows:

$ \frac{\partial\varphi^2}{\partial t^2} -\frac{\partial\varphi^2}{\partial x^2}+\sin \varphi =0, \quad (1)$

and the breather solutions we will study have the form:

$ \varphi (x,t)=4\arctan \left({\frac {{\sqrt {1-\omega ^{2}}}\;\cos(\omega t)}{\omega \;\cosh({\sqrt {1-\omega ^{2}}}\;x)}}\right), \quad (2)$

with $\omega$ a positive constant with a value less than 1. In the coursework you should set $\omega=0.5$ unless otherwise specified.

Eq. (2) (the breather solution) can be used to construct initial data for the initial value problem for Eq. (1) (the PDE). If you correctly implement the time evolution of this data, you should find that it is stable - the evolution is described exactly by the breather solution, which oscillates in time but with the envelope of the function remaining stable. This provides a very good test that the code is working correctly, meaning that you can then confidently use it for different initial data for the same PDE, for which no exact solution exists. Implementing known solutions is a very good way to test a numerical code.

The following analytic results for the breather solution in Eq. (2) may be useful:

Let $A = \sqrt {1-\omega ^{2}}$, then

$ \frac{\partial\varphi (x,t)}{\partial x}= \frac{-8\omega A^2 \cos(\omega t)\sinh(Ax)}
{2 (A \cos(\omega t))^2 +  \omega^2 +  \omega^2 \cosh{2Ax}} \quad (3)$

$ \frac{\partial\varphi (x,t)}{\partial t}= \frac{-4 A \sin(\omega t)}
{\cosh{Ax} \left(1 + \left(\frac{A \cos{wt}}{\omega \cosh{Ax}}\right)^2 \right)} \quad (4)$

**Q1a**. Write a class that implements the breather solution in Eq. (2) above. Think about the appropriate attributes and methods that such a class should have - it may help you to read forward through the instructions to see what methods may be needed. You can also always come back and add more methods/attributes here if you find you need them at a later date.

   As a minimum, you should have methods to return the values of the function at a given time, and for a given set of spatial coordinates, and its time and spatial derivatives. Add a plot to demonstrate how these functions work.

```python
class Breather:
    '''
    Class for the sine Gordon breather solution.

    Attributes:
        w (float): parameter
    '''

    def __init__(self, w=0.5):
        self.w = w
        assert self.w > 0 and self.w < 1, "w must be between 0 and 1"

    @property
    def w(self):
        return self._w
    
    @w.setter
    def w(self, w):
        self._w = w
        assert self.w > 0 and self.w < 1, "w must be between 0 and 1"
        
    # Returns the solution at a given points (x, t)
    def get_solution(self, x, t):
        phi = 4 * np.arctan(np.sqrt(1-self.w**2) * np.cos(self.w * t) / 
                            (self.w * np.cosh(np.sqrt(1-self.w**2) * x)))
        
        return phi

    # Returns the solution as a sympy expression
    def sympy_solution(self):
        x, t, w = sp.symbols('x t w')
        phi = 4 * sp.atan(sp.sqrt(1-w**2) * sp.cos(w * t) / (w * sp.cosh(sp.sqrt(1-w**2) * x)))
        return phi
    
    # Returns dphi/dx at a given points (x, t)
    def get_dphi_dx(self, x, t):
        A = np.sqrt(1-self.w**2)
        dphi_dx = (-8 * self.w * A**2 * np.cos(self.w*t) * np.sinh(A*x) / 
                   (2 * (A * np.cos(self.w*t)) ** 2 + self.w ** 2 + self.w ** 2 * np.cosh(2 * A * x)))
        return dphi_dx
    
    # Returns dphi/dt at a given points (x, t)
    def get_dphi_dt(self, x, t):
        A = np.sqrt(1-self.w**2)
        dphi_dt = (-4 * A * np.sin(self.w*t) / 
                   (np.cosh(A*x) * (1 + (A * np.cos(self.w*t) /( self.w * np.cosh(A*x))) ** 2)))
        return dphi_dt

# I define an instance only once as there is no need to change w
breather = Breather(w=0.5)
```

```python
x = np.linspace(-10, 10, 1000)
y = breather.get_solution(x, 0)
dy_dx = breather.get_dphi_dx(x, 0)
dy_dt = breather.get_dphi_dt(x, 0)

print("Sympy function:\n")
pprint(breather.sympy_solution())

plt.plot(x, y, label='phi')
plt.plot(x, dy_dx, label='dphi_dx')
plt.plot(x, dy_dt, label='dphi_dt')
plt.legend()
plt.xlabel('x')
plt.grid()
plt.title('Demonstration')
plt.show()
```

#### 2.2.3. Q3 Differentiation (22 marks for 2 parts (a)-(b)):

The code to generate the pseudospectral derivative matrices for the first and second derivatives of any function $f(x)$ for a grid containing $N$ points *at the Gauss Lobato points* in $x$ is given below for the standard interval $[-1,1]$, and verified on some example functions.

```python
# Code provided for the derivative matrices at Gauss Lobato nodes for the interval [-1,1]
# (For use in part 2.2.3)

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Code provided for the derivative matrices at Gauss Lobato nodes for the interval [-1,1]

def get_pseudospectral_first_derivative_matrix(N) :
    
    indices = np.arange(N+1)
    u_i = np.cos(np.pi * indices / N)
    c_i = np.ones_like(indices)
    c_i[0] = 2.0
    c_i[N] = 2.0

    D_matrix = np.zeros([N+1,N+1])

    for idx_i in indices :
        for idx_j in indices :
            if(idx_i == 0 and idx_j == 0) : 
                D_matrix[idx_i,idx_j] = (2.0 * N * N + 1.0)/6.0
                
            elif (idx_i == N and idx_j == N) :
                D_matrix[idx_i,idx_j] = -(2.0 * N * N + 1.0)/6.0
                
            elif (idx_i == idx_j) :
                D_matrix[idx_i,idx_j] = - u_i[idx_i] / 2.0 / (1.0 - u_i[idx_i] * u_i[idx_i])
                
            else :
                D_matrix[idx_i,idx_j] = (c_i[idx_i] / c_i[idx_j] * (-1)**(idx_i+idx_j) 
                                         / (u_i[idx_i] - u_i[idx_j]))
    return D_matrix

def get_pseudospectral_second_derivative_matrix(N) :
    
    indices = np.arange(N+1)
    u_i = np.cos(np.pi * indices / N)
    c_i = np.ones_like(indices)
    c_i[0] = 2.0
    c_i[N] = 2.0

    D2_matrix = np.zeros([N+1,N+1])

    for idx_i in indices :
        for idx_j in indices :
            if((idx_i == idx_j) and (idx_i == 0 or idx_i == N)) : 
                D2_matrix[idx_i,idx_j] = (N ** 4.0 - 1.0) / 15.0
            
            elif ((idx_i == idx_j) and (idx_i != 0) and (idx_i != N)) :
                D2_matrix[idx_i,idx_j] = (- ((N*N - 1.0) * (1.0 - u_i[idx_i] * u_i[idx_i]) + 3.0) 
                                             / (3.0 * (1.0 - u_i[idx_i] * u_i[idx_i]) ** 2.0))
                 
            elif (idx_i == 0 and idx_j > 0) :
                D2_matrix[idx_i,idx_j] = ((2.0 / 3.0) * ((-1) ** idx_j) / c_i[idx_j] 
                                          * ((2.0 * N * N + 1.0) * (1.0 - u_i[idx_j]) - 6.0)
                                          / (1.0 - u_i[idx_j]) / (1.0 - u_i[idx_j]))
                
            elif (idx_i == N and idx_j < N) :
                D2_matrix[idx_i,idx_j] = ((2.0 / 3.0) * ((-1) ** (N+idx_j)) / c_i[idx_j] 
                                          * ((2.0 * N * N + 1.0) * (1.0 + u_i[idx_j]) - 6.0)
                                          / (1.0 + u_i[idx_j]) / (1.0 + u_i[idx_j]))
                
            else :
                D2_matrix[idx_i,idx_j] = (((-1) ** (idx_i+idx_j)) / c_i[idx_j] 
                                            * (u_i[idx_i]*u_i[idx_i] + u_i[idx_i]*u_i[idx_j] - 2.0)
                                            / (1.0 - u_i[idx_i]*u_i[idx_i]) / (u_i[idx_i] - u_i[idx_j])
                                            / (u_i[idx_i] - u_i[idx_j]))
                
    return D2_matrix
```

```python
# Check this gives appropriate derivatives for a test function:
N = 8
indices = np.arange(N+1)
u = np.cos(np.pi * indices / N)
phi_test = np.exp(-u*u) + np.sin(u)
D = get_pseudospectral_first_derivative_matrix(N)
Dphi_approx = np.dot(D, phi_test)
D2 = get_pseudospectral_second_derivative_matrix(N)
D2phi_approx = np.dot(D2, phi_test)

u_true = np.linspace(-1,1,100)
Dphi_true = - 2.0 * u_true * np.exp(-u_true*u_true) + np.cos(u_true)
D2phi_true = (4.0 * u_true ** 2.0 - 2.0) * np.exp(-u_true*u_true) - np.sin(u_true)

plt.plot(u_true, Dphi_true, '--', label="true first derivative")
plt.plot(u, Dphi_approx, 'o', label="GL first derivative")
plt.plot(u_true, D2phi_true, '--', label="true second derivative")
plt.plot(u, D2phi_approx, 'o', label="GL second derivative")
plt.xlabel("u")
plt.ylabel("derivative")
plt.legend()
plt.grid();
```

*Answer:*

For the finite difference stencils to be second order accurate, we need to have $2 + 1 = 3\space points$ in the stencils for the 1st and 2nd derivatives. To find the coefficients, we must fit a Lagrange polynomial to the three evenly spaced points in the stencil. We must then take the first and second derivatives of this polynomial, and find their values at the central point. This gives us the coefficients for the finite difference stencils. 

For the second-order boudary stencils for x', we can use 3 points as above and evaluate the derivative of the Lagrange polynomial at the edge point. For the second-order boundary stencils for x'', it turns out we need 4 points according to numerous sources online like [this](https://math.unm.edu/~motamed/Teaching/OLD/Fall20/HPSC/fd.html)  (not going to lie but I'm confused as to why this is the case). I implemented this using sympy similar to the other stencils. Alternatively, we can use lower order accuracy stencils, or just set the boundary points derivatives to zero, especially when we are looking at a wide interval like $[-15,15]$.

```python
# 1st and 2nd derivative stencils for 3 points

# Construct the polynomial using sympy
x = symbols('x')
deltax = symbols('dx')
x2 = symbols('x2')
x1 = x2-deltax
x3 = x2+deltax
f1 = symbols('f1')
f2 = symbols('f2')
f3 = symbols('f3')
L1 = (x - x2 ) / (x1 - x2) * (x - x3 ) / (x1 - x3)
L2 = (x - x1 ) / (x2 - x1) * (x - x3 ) / (x2 - x3)
L3 = (x - x1 ) / (x3 - x1) * (x - x2 ) / (x3 - x2)
P = f1 * L1 + f2 * L2 + f3 * L3
print("Using sympy we get the polynomial", simplify(P))

# Taking the derivatives
dPdx = diff(P, x)
dPdx2 = diff(dPdx, x)
print("The first derivative is", simplify(dPdx))
print("The second derivative is", simplify(dPdx2))

# Finding the values of the derivatives at the central point x2
dPdx_at_x2 = dPdx.subs(x, x2)
dPdx2_at_x2 = dPdx2.subs(x, x2)
print("The first derivative at x2 is", simplify(dPdx_at_x2))
print("The second derivative at x2 is", simplify(dPdx2_at_x2))
print()
print("First derivative stencil:")
pprint(simplify(dPdx_at_x2))
print()
print("Second derivative stencil:")
pprint(simplify(dPdx2_at_x2))
```

```python
# Left edge second degree stencil for x''
# Construct the polynomial using sympy
x = symbols('x')
deltax = symbols('dx')
x1 = symbols('x1')
x2 = x1 + deltax
x3 = x1 + 2*deltax
x4 = x1 + 3*deltax
f1 = symbols('f1')
f2 = symbols('f2')
f3 = symbols('f3')
f4 = symbols('f4')
L1 = (x - x2 ) / (x1 - x2) * (x - x3 ) / (x1 - x3) * (x - x4 ) / (x1 - x4)
L2 = (x - x1 ) / (x2 - x1) * (x - x3 ) / (x2 - x3) * (x - x4 ) / (x2 - x4)
L3 = (x - x1 ) / (x3 - x1) * (x - x2 ) / (x3 - x2) * (x - x4 ) / (x3 - x4)
L4 = (x - x1 ) / (x4 - x1) * (x - x2 ) / (x4 - x2) * (x - x3 ) / (x4 - x3)
P = f1 * L1 + f2 * L2 + f3 * L3 + f4 * L4
print("Using sympy we get the polynomial", simplify(P))

# Now we take the derivative
dPdx = diff(P, x)
dPdx2 = diff(dPdx, x)
print("The derivative is", simplify(dPdx2))

# Now we find the value at the left edge x1
dPdx2_at_x1 = dPdx2.subs(x, x1)
print("The second derivative at x1 is", simplify(dPdx2_at_x1))
print("This gives the second-order left edge stencil for x'': \n")
pprint(simplify(dPdx2_at_x1))
```

```python
# Right edge second degree stencil for x''
# Construct the polynomial using sympy
x = symbols('x')
deltax = symbols('dx')
x4 = symbols('x4')
x3 = x4 - deltax
x2 = x4 - 2*deltax
x1 = x4 - 3*deltax
f1 = symbols('f1')
f2 = symbols('f2')
f3 = symbols('f3')
f4 = symbols('f4')
L1 = (x - x2 ) / (x1 - x2) * (x - x3 ) / (x1 - x3) * (x - x4 ) / (x1 - x4)
L2 = (x - x1 ) / (x2 - x1) * (x - x3 ) / (x2 - x3) * (x - x4 ) / (x2 - x4)
L3 = (x - x1 ) / (x3 - x1) * (x - x2 ) / (x3 - x2) * (x - x4 ) / (x3 - x4)
L4 = (x - x1 ) / (x4 - x1) * (x - x2 ) / (x4 - x2) * (x - x3 ) / (x4 - x3)
P = f1 * L1 + f2 * L2 + f3 * L3 + f4 * L4
print("Using sympy we get the polynomial", simplify(P))

# Now we take the derivative
dPdx = diff(P, x)
dPdx2 = diff(dPdx, x)
print("The second derivative is", simplify(dPdx2))

# Now we find the value at the right edge x4
dPdx2_at_x4 = dPdx2.subs(x, x4)
print("The second derivative at x4 is", simplify(dPdx2_at_x4))
print("This gives the second-order right edge stencil for x'': \n")
pprint(simplify(dPdx2_at_x4))
```

Now we construct the matrices

```python
def get_fds_first_derivative_matrix(N):
    
    indices = np.arange(N+1)

    dx = 2.0 / N

    D_matrix = np.zeros([N+1,N+1])

    for idx_i in indices :
        for idx_j in indices :
            # Central points - central difference stencil
            if idx_i == idx_j:
                D_matrix[idx_i,idx_j] = 0

            elif idx_i == idx_j + 1:
                D_matrix[idx_i,idx_j] = - 1 / (2*dx)

            elif idx_i == idx_j - 1:
                D_matrix[idx_i,idx_j] = 1 / (2*dx)

            else:
                D_matrix[idx_i,idx_j] = 0
            
    # Boundary points
    D_matrix[0, 0] = -3 / (2*dx)
    D_matrix[0,1] = 4 / (2*dx)
    D_matrix[0,2] = -1 / (2*dx)
    D_matrix[N,N] = 3 / (2*dx)
    D_matrix[N,N-1] = -4 / (2*dx)
    D_matrix[N,N-2] = 1 / (2*dx)

    return D_matrix 

def get_fds_second_derivative_matrix(N):    

    indices = np.arange(N+1)

    dx = 2.0 / N

    D2_matrix = np.zeros([N+1,N+1])

    for idx_i in indices :
        for idx_j in indices :
            # Central points - central difference stencil
            if idx_i == idx_j -1:
                D2_matrix[idx_i,idx_j] = 1 / dx**2
            elif idx_i == idx_j:
                D2_matrix[idx_i,idx_j] = -2 / dx**2
            elif idx_i == idx_j + 1:
                D2_matrix[idx_i,idx_j] = 1 / dx**2
            else:
                D2_matrix[idx_i,idx_j] = 0
        
    # Boundary points
    D2_matrix[0,0] = 2 / dx**2
    D2_matrix[0,1] = -5 / dx**2
    D2_matrix[0,2] = 4 / dx**2
    D2_matrix[0,3] = -1 / dx**2
    D2_matrix[N,N] = 2 / dx**2
    D2_matrix[N,N-1] = -5 / dx**2
    D2_matrix[N,N-2] = 4 / dx**2
    D2_matrix[N,N-3] = -1 / dx**2

    return D2_matrix
```

**Q4a**. Explain what type of PDE Eq. (1) is and discuss how this impacts on well-posedness and the way in which we can solve the equation numerically.

*Answer*

The Sine-Gordon equation is a hyperbolic PDE. On top of that it is symmetric hyperbolic, which can be shown if we rewrite the equation similar to the wave function:
$$\frac{\partial}{\partial t}\begin{bmatrix}u\\ v\\ \psi \end{bmatrix} = \begin{bmatrix}0 & 0 & 0\\ 0 & 0 & 1\\ 0 & 1 & 0\end{bmatrix}\frac{\partial}{\partial x}\begin{bmatrix}u\\ v\\ \psi \end{bmatrix} + \begin{bmatrix}v \\ -sin(u) \\ 0\end{bmatrix}$$

,where

$$\psi = \frac{\partial u}{\partial x} \qquad and \qquad v = \frac{\partial u}{\partial t}$$

The eigenvalues of the matrix M are $\lambda_1 = 1, \lambda_2 = -1, \lambda_3 = 0$, but we can also show that $M = (M^T)^{*}$ which means that the Sine-Gordon equation is symmetric hyperbolic and therefore well-posed.

It is also important to note that the Sine-Gordon equation is a nonlinear PDE because of the $sin(\phi)$. 

**Q4b**. Solve the equation using finite differences with a fixed grid spacing and the explicit midpoint/RK2 time integration method. Discuss the CFL condition and how it applies here. Confirm that the solution remains stationary (although oscillating in time).
   
_(HINT: you can reuse the derivative matrices from part 2.2.3!)_

*Answer:*

Here I implemented the code from the courseworks almost exactly. The most notable change comes in calculate_dydt, where we need to subtract $sin(u)$ from the approximated $\frac{\partial u^2}{\partial ^2 x}$ to find $\frac{\partial v}{\partial t}$. I would have wished to do it myself, however, time was scarce. 

I take an interval from -10 to 10 over which to integrate the PDE. If the interval is too small we might run into issues with setting the boundary derivatives to 0 and therefore I'm looking at $[-10, 10]$.

```python
class SineGordonPDE :
    # constructor function
    def __init__(self, a_x, a_y0, a_max_time, a_method="finite_difference"):
        self.max_time = a_max_time  # maximum time for integration
        self.x = a_x  # grid points in x direction
        self.y0 = a_y0  # initial condition
        self.N_grid = np.size(a_x)  # number of grid points
        self._solution_t = []  # time solution
        self._solution_y = []  # y solution
        self.method = a_method  # method for time integration
        self.breather = Breather(w=0.5)  # breather object
        
        self.initial_value = self.y0  # initial value of y
        
        if self.method == "finite_difference" :
            self.D2_matrix = get_fds_second_derivative_matrix(self.N_grid-1)  # finite difference second derivative matrix
        elif self.method == "pseudospectral" :
            self.D2_matrix = get_pseudospectral_second_derivative_matrix(self.N_grid-1)  # pseudospectral second derivative matrix
        else :
            assert False, "Requested derivative method not implemented"
            
        # Rescale for the derivative matrix for the interval we are looking at
        b_minus_a = self.x[self.N_grid-1] - self.x[0]  # interval length
        self.D2_matrix *= 4.0 / b_minus_a / b_minus_a  # rescale the derivative matrix

    def calculate_dydt(self, t, current_state) :
        # Split the current state into u and v
        [u,v] = np.array_split(current_state, 2)
        dydt = np.zeros_like(current_state)  # initialize the time derivative
        dudt, dvdt = np.array_split(dydt, 2)  # split the time derivative
        
        # Calculate the time derivatives
        dudt[:] =  v  # du/dt = v
        dvdt[:] =  np.dot(self.D2_matrix, u) - np.sin(u)  # dv/dt = D2(u) - sin(u)
           
        # Zero the derivatives at the end for stability 
        dydt[0] = 0.0
        dydt[1] = 0.0
        dydt[self.N_grid-1] = 0.0
        dydt[self.N_grid-2] = 0.0
        dydt[self.N_grid] = 0.0
        dydt[self.N_grid+1] = 0.0
        dydt[self.N_grid*2-1] = 0.0
        dydt[self.N_grid*2-2] = 0.0
        
        return dydt      
 
    def integrate_using_solve_ivp(self, a_num_steps=100) :
        # Define the time points for the solution
        t_solution = np.linspace(0.0, self.max_time, a_num_steps)
        delta_t = self.max_time / (a_num_steps - 1)
        
        # Integrate the PDE using solve_ivp
        solution = solve_ivp(self.calculate_dydt, [0, self.max_time],
                        self.initial_value, t_eval=t_solution)
        
        self._solution_t = solution.t  # store the time solution
        self._solution_y = solution.y  # store the y solution
        
        self._solution_solved = True  # mark the solution as solved
        
        return solution.t, solution.y
    
    def integrate_using_midpoint(self, a_CFL=0.5) :
        # assert self.method == "finite_difference", 'need to use FDs'
        dx = self.x[1] - self.x[0]  # grid spacing
        
        # Set the time step
        delta_t = a_CFL * dx
        num_timesteps = int(self.max_time / delta_t) + 1
        t_solution = np.linspace(0.0, self.max_time, num_timesteps)
        y_solution = np.zeros([2*self.N_grid, num_timesteps])
        
        for itime, time in enumerate(t_solution) :
            if itime == 0 :
                y_solution[:, itime] = self.initial_value
                old_time = time
            else :
                dydt_at_t = self.calculate_dydt(old_time, y_solution[:, itime-1])
                
                # Take half step
                y_t_plus_half = y_solution[:, itime-1] + 0.5 * dydt_at_t * delta_t
                t_plus_half = old_time + 0.5 * delta_t
                dydt_at_t_plus_half = self.calculate_dydt(t_plus_half, y_t_plus_half)
                
                # Take full step
                y_solution[:, itime] = (y_solution[:, itime-1] + dydt_at_t_plus_half * delta_t) 
                old_time = time 

        self._solution_t = t_solution
        self._solution_y = y_solution
        
        return t_solution, y_solution
    
    def integrate_using_implicit_method(self, a_num_steps=1000) :
        
        t_solution = np.linspace(0.0, self.max_time, a_num_steps)
        y_solution = np.zeros([2*self.N_grid, a_num_steps])
        delta_t = self.max_time / (a_num_steps - 1)
        
        for itime, time in enumerate(t_solution) :
            if itime == 0 :
                y_solution[:,itime] = self.initial_value
            else :
                y_of_t_old   = y_solution[:,itime-1]
                
                y_of_t_guess = np.copy(y_of_t_old)
                
                # Iterative scheme from letcure notes
                error = 10.0
                error_threshold = 1e-1
                while (error > error_threshold) :
                    
                    # Get the value of dydt at the guessed value
                    dydt_at_t_next = self.calculate_dydt(time, y_of_t_guess)
                    dydt_at_t_old  = self.calculate_dydt(time-1, y_of_t_old)
                    
                    # Get the new guess using the Trapezoid update rule
                    y_of_t_new = y_of_t_old + delta_t / 2 * (dydt_at_t_next + dydt_at_t_old)
                    
                    # If the new guess is the same as the old one we have converged
                    error = np.linalg.norm(y_of_t_guess - y_of_t_new)
                    y_of_t_guess = y_of_t_new
                    
                    # Check it is not getting worse!
                    assert error < 20.0, 'Non linear iterations not converging!'
                    
                # Store the solution
                y_of_t_guess[0] = 0.0
                y_of_t_guess[self.N_grid-1] = 0.0
                y_solution[:,itime] = y_of_t_guess
                
        # Store the solution
        self._solution_t = t_solution
        self._solution_y = y_solution
        
        return t_solution, y_solution
```

```python
# Test the solution
num_points = 100
a=-10
b=10
x_values = np.linspace(a , b,  num_points)
u0 = breather.get_solution(x_values, 0)
v0 = breather.get_dphi_dt(x_values, 0)
y0 = np.concatenate([u0,v0])
max_time = 10

my_wave_equation = SineGordonPDE(x_values, y0, max_time)
t_solution, y_solution = my_wave_equation.integrate_using_midpoint(a_CFL=0.25)

u_solution = y_solution[0:num_points,:]

for i, t_i in enumerate(t_solution) :
    if (i > 0) and (i % 3 == 0) and (t_i > 0.0):
        labelt = "t="+str(round(t_i,2))
        f_t = u_solution[:,i]
        plt.plot(x_values, f_t, alpha=(t_i/max_time))

plt.title("Sine Gordon breather solution")
plt.xlabel('x')
plt.ylabel('value over time of u')
plt.grid()
```

*Answer:*

For the midpoint method to work, we need to lower the CFL value from 1 to 0.5 or to 0.25. This is because the midpoint method is second order accurate, so we want to take more steps to get a more accurate solution. When CFL=1 and we take a large interval to integrate over, the steps become very big and the results inaccuarate with the boundary points exploding. Lowering the CFL value makes our solution stationary. From the plot we can see that even though the solution is oscillating in time, the envelope of the function remains stable.
